```python
import openpyxl

# Загрузка Excel-документа
file_path = 'part.xlsx'
workbook = openpyxl.load_workbook(file_path)
sheet = workbook.active

# Создаем список для хранения результатов
results = []

# Проходим по всем строкам в столбце 13 (M)
for row in sheet.iter_rows(min_row=1, max_col=13, max_row=sheet.max_row):
    cell = row[12]  # Индекс 12 соответствует столбцу M (13-й столбец)
    
    # Проверяем, есть ли в ячейке текст и гиперссылка
    cell_text = cell.value
    cell_hyperlink = cell.hyperlink.target if cell.hyperlink else None
    
    # Добавляем результаты в список
    results.append((cell_text, cell_hyperlink))

# Выводим результаты
for text, hyperlink in results:
    print(f"Текст: {text}, Гиперссылка: {hyperlink}")

# Закрываем файл
workbook.close()
```

```python
import pandas as pd

# Загрузка данных из Excel файла
df = pd.read_excel('doc.xlsx')

# Инициализация списка для хранения названий столбцов с проблемами
columns_with_issues = []

# Проход по каждому столбцу в DataFrame
for column in df.columns:
    # Проверка, есть ли в столбце значения, превышающие 255 символов
    if df[column].apply(lambda x: len(str(x)) > 255 if pd.notnull(x) else False).any():
        columns_with_issues.append(column)

# Вывод названий столбцов с проблемами
print("Столбцы с значениями, превышающими 255 символов:", columns_with_issues)
```
