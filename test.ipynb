```python
import pandas as pd
import re
import psycopg2

# Путь к файлу Excel
excel_file_path = 'part.xlsx'

# Чтение данных из Excel, начиная с 5-й строки
df = pd.read_excel(excel_file_path, skiprows=4)


# Выбор нужных столбцов
dff = df.iloc[:, [1, 2,3,4]]



# Применение функции к столбцу 'name'
dff[dff.columns[0]] = df[dff.columns[0]].str.replace(r" тип \d+", "", regex=True)
dff = dff.drop_duplicates(subset=dff.columns[0])
dff
# dff.to_excel('part2.xlsx', index=False)
```
