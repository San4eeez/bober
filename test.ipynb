```python
import openpyxl

# Загрузка Excel-документа
file_path = 'part.xlsx'
workbook = openpyxl.load_workbook(file_path)
sheet = workbook.active

# Создаем список для хранения результатов
results = []

# Проходим по всем строкам в столбце 13 (M)
for row in sheet.iter_rows(min_row=1, max_col=14, max_row=sheet.max_row):
    cell = row[13]  # Индекс 12 соответствует столбцу M (13-й столбец)
    
    # Проверяем, есть ли в ячейке текст и гиперссылка
    cell_text = cell.value
    cell_hyperlink = cell.hyperlink.target if cell.hyperlink else None
    
    # Добавляем результаты в список
    results.append((cell_text, cell_hyperlink))

# Выводим результаты
for text, hyperlink in results:
    print(f"Текст: {text}, Гиперссылка: {hyperlink}")

# Закрываем файл
workbook.close()
```

```python
from openpyxl import load_workbook

# Загрузка файла
file_path = 'part.xlsx'
workbook = load_workbook(filename=file_path)
sheet = workbook.active  # или укажите конкретный лист: workbook['SheetName']

# Проход по 14-му столбцу (это столбец 'N' в Excel)
column_number = 14
for row in sheet.iter_rows(min_col=column_number, max_col=column_number, values_only=False):
    cell = row[0]  # Получаем ячейку
    if cell.hyperlink:  # Проверяем наличие гиперссылки
        full_hyperlink = cell.hyperlink.target
        # Проверяем, содержит ли гиперссылка '#' и выводим полную гиперссылку
        if '#' in full_hyperlink:
            print(f"Текст: {cell.value}, Полная гиперссылка: {full_hyperlink}")
        else:
            print(f"Текст: {cell.value}, Гиперссылка: {full_hyperlink}")
    else:
        print(f"Текст: {cell.value}, Гиперссылка: отсутствует")

workbook.close()
```
